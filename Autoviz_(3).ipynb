#Seaborn
<p>Seaborn provides several built-in datasets, including "iris", "penguins", "flights", "titanic", "tips", "diamonds", and "exercise", which can be loaded using the seaborn.load_dataset() function. </p>

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips') #tips is available in seaborn
```

```python
df.head()
```

```python
from autoviz.AutoViz_Class import AutoViz_Class
```

AV=autoViz_clss()

```python
AV = AutoViz_Class()
dft = AV.AutoViz(
    "", #this is for a filename if you aren't using a dataframe
    sep=",", #separator in the file, generally comma
    depVar="size", #this is for a target variable name if you have one that you are using in prediction
    dfte=df, #dataframe that has been used for the file that has been input
    header=0, #header row which is typically the first
    verbose=1, #0 minimal charts,  1 is more charts, 2 is for saving locally without display
    lowess=True, #creates regression line with dependent variable for all continuous variables
    chart_format="bokeh",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    save_plot_dir=None
)
```

```python


df2 = sns.load_dataset("penguins").dropna()

AV2 = AutoViz_Class()
dft2 = AV2.AutoViz(
    "",
    sep=",",
    depVar="body_mass_g",
    dfte=df2,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="bokeh",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    save_plot_dir=None
)
```

#from AutoViz documentation in Github
<p
Arguments for AV.AutoViz() method:<br>

filename: Use an empty string ("") if there's no associated filename and you want to use a dataframe. In that case, using the dfte argument for the dataframe. Otherwise provide a filename and leave dfte argument with an empty string. Only one of them can be used.<br>
sep: File separator (comma, semi-colon, tab, or any column-separating value) if you use a filename above.<br>
depVar: Target variable in your dataset; set it as an empty string if not applicable.<br>
dfte: name of the pandas dataframe for plotting charts; leave it as empty string if using a filename.<br>
header: set the row number of the header row in your file (0 for the first row). Otherwise leave it as 0.<br>
verbose: 0 for minimal info and charts, 1 for more info and charts, or 2 for saving charts locally without display.<br>
lowess: Use regression lines for each pair of continuous variables against the target variable in small datasets; avoid using for large datasets (>100,000 rows).<br>
chart_format: 'svg', 'png', 'jpg', 'bokeh', 'server', or 'html' for displaying or saving charts in various formats, depending on the verbose option.<br>
max_rows_analyzed: Limit the max number of rows to use for visualization when dealing with very large datasets (millions of rows). A statistically valid sample will be used by autoviz. Default is 150000 rows.<br>
max_cols_analyzed: Limit the number of continuous variables to be analyzed. Defaul is 30 columns.<br>
save_plot_dir: Directory for saving plots. Default is None, which saves plots under the current directory in a subfolder named AutoViz_Plots. If the save_plot_dir doesn't exist, it will be created.</p>
